### Imports and constants

```python
from google.colab import userdata
from huggingface_hub import login

import torch
from transformers import pipeline
from diffusers import DiffusionPipeline, AutoPipelineForText2Image
from IPython.display import display
from PIL import Image
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
### Constants
MODEL = 'stabilityai/stable-diffusion-xl-base-1.0'
PROMPT = 'Generate an image representing seaside town like St. Ives and Newquay, \
painted in a style of pop-art, featuring visible brushstrokes, vibrant colours for blue sky, \
a few surfers, some wild rocks on shore, beaches and white waves'
SIZE = '1024x1024'
```

```python
hf_token = userdata.get('HF_TOKEN')
login(hf_token, add_to_git_credential=True)
```

### Image generation using DissusionPipeline

```python
%%time
pipeline = DiffusionPipeline.from_pretrained(
    pretrained_model_name_or_path=MODEL,
    torch_dtype=torch.float16,
    use_safetensors=True,
    variant='fp16').to('cuda')

image = pipeline(prompt=PROMPT).images[0]
```

```python
display(image)
plt.figure(figsize=(10,10), dpi=200)
plt.imshow(X=image)
plt.axis('off')
plt.show()
```

![seaside](seaside1.png)

```python
# Save the image
image.save('seaside1.png')
```

```python
del pipeline, image
torch.cuda.empty_cache()
```

### Image generation using AutoPipeline

```python
%%time
pipeline = AutoPipelineForText2Image.from_pretrained(
    pretrained_model_or_path=MODEL,
    torch_dtype=torch.float16,
    use_safetensors=True).to('cuda')

image = pipeline(prompt=PROMPT,
    guidance_scale=7.5,
    variant='fp16').images[0]
```

```python
# Display
plt.figure(figsize=(10,10), dpi=200)
plt.imshow(X=image)
plt.axis('off')
plt.show()
```
